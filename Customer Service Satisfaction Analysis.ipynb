### I. GIỚI THIỆU: 
Bộ dữ liệu ghi lại điểm hài lòng của khách hàng trong khoảng thời gian một tháng tại nền tảng thương mại điện tử có tên Shopzilla. Nó bao gồm trường khác nhau như danh mục và danh mục phụ, nhận xét của khách hàng, ngày phản hồi khảo sát, giá mặt hàng, chi tiết đại lý (tên, người giám sát, người quản lý) và điểm hài lòng của khách hàng (CSAT),...

CSAT là thước đo dùng để đánh giá mức độ hài lòng của khách hàng đối với sản phẩm, dịch vụ do doanh nghiệp cung cấp. Trong bối cảnh thương mại điện tử, CSAT đóng một vai trò quan trọng trong việc đo lường mức độ công ty đáp ứng mong đợi và nhu cầu của khách hàng, cuối cùng ảnh hưởng đến khả năng giữ chân và danh tiếng của thương hiệu.

- Khám phá xu hướng điểm CSAT.
Mục tiêu chính của báo cáo này là phân tích các xu hướng trong bộ dữ liệu:
- Trực quan hóa mối liên hệ giữa các yếu tố và điểm CSAT.
- Phân tích điểm số giữa các nhân viên tiếp nhận phản hồi, người giám sát, hạng mục...
- Xác định các yếu tố ảnh hưởng đến điểm CSAT.
- Đưa ra các đề xuất nhằm nâng cao sự hài lòng của khách hàng.

#### Dưới đây là giải thích ý nghĩa của mỗi trường thông tin:

- Unique id: Đây là một ID duy nhất được gán cho mỗi bản ghi trong bộ dữ liệu, giúp xác định mỗi bản ghi một cách duy nhất.

- channel_name: Đây là tên của kênh liên lạc mà khách hàng sử dụng để liên hệ hoặc gửi yêu cầu. Ví dụ: Inbound (gọi đến), Outbound (gọi đi), Email,.. 

- category: Đây là danh mục chính của vấn đề hoặc loại dịch vụ mà khách hàng đang gặp phải hoặc yêu cầu hỗ trợ.

- Sub-category: Đây là các loại con hoặc chi tiết hơn của các vấn đề được phân loại trong trường 'category'.

- Customer Remarks: Đây là ghi chú hoặc nhận xét từ phía khách hàng liên quan đến vấn đề hoặc yêu cầu hỗ trợ của họ.

- Order_id: Đây là ID của đơn hàng hoặc giao dịch liên quan đến vấn đề hoặc yêu cầu dịch vụ.

- order_date_time: Đây là thời gian và ngày mà đơn hàng hoặc giao dịch đã được tạo ra hoặc thực hiện.

- Issue_reported at: Đây là thời gian và ngày mà vấn đề hoặc yêu cầu dịch vụ đã được báo cáo.

- issue_responded: Đây là thời gian và ngày mà vấn đề hoặc yêu cầu dịch vụ đã được phản hồi hoặc xử lý.

- Survey_response_Date: Đây là thời gian và ngày mà khách hàng đã phản hồi hoặc hoàn thành cuộc khảo sát (nếu có).

- Customer_City: Đây là thành phố hoặc địa điểm nơi khách hàng đang hoặc đã tạo ra vấn đề hoặc yêu cầu dịch vụ.

- Product_category: Đây là danh mục của sản phẩm liên quan đến vấn đề hoặc yêu cầu dịch vụ.

- Item_price: Đây là giá của sản phẩm hoặc dịch vụ liên quan đến vấn đề hoặc yêu cầu dịch vụ.

- connected_handling_time: Đây là thời gian xử lý hoặc thời gian liên kết trong quá trình xử lý vấn đề hoặc yêu cầu dịch vụ.

- Agent_name: Đây là tên của nhân viên hoặc đại diện của tổ chức hoặc doanh nghiệp đang xử lý vấn đề hoặc yêu cầu dịch vụ.

- Supervisor, Manager: Đây là tên của người giám sát hoặc quản lý trực tiếp của nhân viên hoặc đại diện đang xử lý vấn đề hoặc yêu cầu dịch vụ.

- Tenure Bucket: Đây là phân loại thời gian làm việc của nhân viên hoặc đại diện theo khoảng thời gian nhất định.

- Agent Shift: Đây là ca làm việc của nhân viên hoặc đại diện đang xử lý vấn đề hoặc yêu cầu dịch vụ.

- CSAT Score: Đây là điểm đánh giá CSAT (Customer Satisfaction Score), thường được khách hàng đánh giá để đo lường mức độ hài lòng của họ về dịch vụ hoặc hỗ trợ họ nhận được.

### Phân tích bao gồm các yếu tố như kênh, danh mục, danh mục phụ, nhận xét của khách hàng, chi tiết đơn hàng, thời gian xử lý, nhân viên trực , người giám sát, người quản lý, ca làm việc và thành phố.

```python
#Chuyển đổi định dạng của các trường: order_date_time, Issue_reported at, issue_responded, Survey_response_Date thành ngày giờ.
csat['order_date_time'] = pd.to_datetime(csat['order_date_time'], format="%d/%m/%Y %H:%M")
csat['Issue_reported at'] = pd.to_datetime(csat['Issue_reported at'], format="%d/%m/%Y %H:%M")
csat['Issue_reported_date'] = pd.to_datetime(csat['Issue_reported at'], format="%d/%m/%Y %H:%M").dt.date
csat['issue_responded'] = pd.to_datetime(csat['issue_responded'], format="%d/%m/%Y %H:%M")
csat['Survey_response_Date'] = pd.to_datetime(csat['Survey_response_Date'], format="%d-%b-%y")
```

```python
#Tính điểm CSAT trung bình mỗi ngày
csat_time = csat.groupby('Issue_reported_date')['CSAT Score'].mean().reset_index(name='Score_avg')
csat_time.head()
```

```python
#Biểu đồ thể hiện trung bình điểm CSAT qua các ngày
plt.figure(figsize=(12, 6))
sns.lineplot(data=csat, x='Issue_reported_date', y='CSAT Score')
plt.xlabel('Issue Reported Date')
plt.ylabel('CSAT Score')
plt.title('CSAT Score over Time')
plt.xticks(csat_time['Issue_reported_date'],rotation=60) 
plt.tight_layout()
plt.show()
```

#### Nhận xét: 
Điểm CSAT đột ngột xuống thấp vào khoảng 2,5 ở ngày 29/07/2023, sau đó tăng trở lại đến ngày 31/07/2023 và luôn ổn định ở mức hơn 4, cho thấy khách hàng khá hài lòng với các vấn đề được giải quyết trong tháng 8/2023.

```python
#Biểu đồ phân bố điểm CSAT trong tổng số đánh giá ghi nhận được
plt.figure(figsize=(10,5))
sns.set_palette("Set2")


plt.subplot(1,2,1)
sns.barplot(data=csat_score_counts, x='CSAT Score', y='Count');
for index, row in csat_score_counts.iterrows():
    plt.text(index, row['Count']+1000, str(row['Count']), ha='center')
plt.ylabel('Total reports')
plt.title('Distribution of CSAT Scores')

    
plt.subplot(1,2,2)
plt.pie(csat_score_counts['Count'], labels=csat_score_counts['CSAT Score'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of CSAT Scores')
plt.legend(loc='center right')
plt.tight_layout()
plt.show()
```

#### Nhận xét: 
Đánh giá 5 điểm chiếm tỷ trọng rất cao gần 70% trong tổng số đánh giá được ghi nhận. 
Số đánh giá 1 điểm và 4 điểm gần bằng nhau (13,1%). 
Chiếm tỷ lệ rất ít và thấp nhất lần lượt là 2 điểm (3%) và 3 điểm (1,5%)

```python
#Biểu đồ phân bố điểm CSAT theo người quản lý
plt.figure(figsize=(10,5))
sns.set_palette("Set2")

plt.subplot(1,2,1)
sns.barplot(data=csat_distribute_byManager, x='Manager', y='Count')
for index, row in csat_distribute_byManager.iterrows():
    plt.text(row.name, row['Count']+100, str(row['Count']), ha='center')
plt.xticks(rotation=60)
plt.ylabel('Total reports')

plt.subplot(1,2,2)
plt.pie(csat_distribute_byManager['Count'], labels=csat_distribute_byManager['Manager'], autopct='%1.1f%%', startangle=140)
plt.title('CSAT Distribution by Manager')
plt.show()
```

```python
# Thống kê điểm đánh giá trung bình theo từng người quản lý
csat_score_byManager = csat.groupby('Manager')['CSAT Score'].mean()
csat_score_byManager = pd.DataFrame({'Manager': csat_score_byManager.index, 'Score_avg': csat_score_byManager.values})
csat_score_byManager = csat_score_byManager.sort_values(by='Score_avg').reset_index(drop=True)
csat_score_byManager
```

```python
# Biểu đồ thể hiện điểm đánh giá trung bình theo từng người quản lý.
sns.set_palette("Set2")

sns.barplot(data=csat_score_byManager, x='Manager', y='Score_avg', order=csat_score_byManager['Manager']);
plt.xticks(rotation=60);
for index, row in csat_score_byManager.iterrows():
    plt.annotate(f'{row["Score_avg"]:.2f}', (index, row['Score_avg']), ha='center', va='bottom')
plt.title('CSAT average by Manager');
plt.ylabel('CSAT score average');
```

#### Nhận xét:
Điểm đánh giá trung bình ghi nhận được theo người quản lý tương đối cao và đồng đều nhau, dao động trong khoảng 4,11 (quản lý Olivia Tan) đến 4,38 (quản lý Emily Chen).

```python
#Thống kê sự phân bổ số lượng đánh giá theo nguyên nhân
csat_distribute_byCategory = csat.groupby('category').size()
csat_distribute_byCategory = pd.DataFrame(csat_distribute_byCategory, columns=['Count']).reset_index()
csat_distribute_byCategory
```

```python
# Thống kê điểm số CSAT trung bình theo danh mục vấn đề
                                                                
csat_score_byCategory = csat.groupby('category')['CSAT Score'].mean().round(2).reset_index(name='Score_avg')
csat_score_byCategory= csat_score_byCategory.sort_values(by='Score_avg').reset_index(drop=True)
csat_score_byCategory
```

```python
# Thống kê điểm số CSAT trung bình theo danh mục phụ 

csat_byCategory=pd.merge(csat_distribute_byCategory,csat_score_byCategory, how='inner' )
csat_byCategory=csat_byCategory.sort_values(by='Count').reset_index(drop=True)
csat_byCategory
```

```python
# Biểu đồ thể hiện điểm số CSAT trung bình theo danh mục vấn đề
plt.figure(figsize=(10, 6))


plt.bar(csat_byCategory['category'], csat_byCategory['Count'], color='blue', label='Total reports');
plt.xticks(rotation=45);
plt.ylabel('Total reports');
plt.legend(loc='upper left');

plt.twinx()
plt.plot(csat_byCategory['category'], csat_byCategory['Score_avg'], color='red', marker='o', label='Score average');
plt.title('CSAT by Category');
plt.xlabel('Category');
plt.ylabel('Score average');
plt.legend(loc='upper right');
```

#### Nhận xét:
Đánh giá của khách hàng trong danh mục vấn đề do "hoàn trả" (return) chiếm số lượng nhiều nhất 44097 lượt, dù vậy trung bình điểm đánh giá vẫn ở mức gần như cao nhất (4,35 điểm) trong các danh mục. Nguyên nhân khách hàng phản ánh trong danh mục "liên quan đến đặt hàng" chiếm khá nhiều nhưng trung bình điểm đánh giá thấp (4,1).

Các category nhận được điểm trung bình từ khách hàng khá cao, trong khoảng từ 3.43 đến 4.40. Điều này cho thấy rằng phần lớn các trường hợp được xử lý đều đạt được sự hài lòng từ phía khách hàng. "App/website" có điểm trung bình cao nhất (4.40), có thể thấy hệ thống ứng dụng và website của công ty có độ ổn định và sự dễ sử dụng tốt.

```python
# Tính thời gian phản hồi của công ty đến các vấn đề từ khách hàng
csat['responded_time'] = pd.to_datetime(csat['issue_responded']) - pd.to_datetime(csat['Issue_reported at'])
csat['responded_time']=csat['responded_time'].dt.total_seconds() / 3600

#Loại bỏ những hàng có thời gian 'issue_responded' bị ghi nhận sai 
csat_new1 = csat[~(csat['issue_responded'].dt.time == pd.to_datetime('00:00:00').time())]
csat_new1.sort_values(by='responded_time')
```

```python
# Biểu đồ thể hiện mối tương quan giữa điểm đánh giá và thời gian phản hồi của công ty đến các vấn đề của khách hàng

sns.scatterplot(csat_new1,x='CSAT Score', y='responded_time');
plt.title('Correlation between CSAT scores and issue response time');
plt.ylabel('Responded time');
```

#### Nhận xét:
Qua biểu đồ phân tán trên, ta thấy điểm đánh giá CSAT và thời gian phản hồi khách hàng không có mối quan hệ tương quan với nhau. 
Thời gian phản hồi dài hay ngắn đều có khả năng được đánh giá cao hoặc bị đánh giá thấp. 

```python
# Loại bỏ những bản ghi không điền giá sản phẩm 
csat_new2=csat_new2 = csat.dropna(subset=['Item_price'])
csat_new2.head()
```
